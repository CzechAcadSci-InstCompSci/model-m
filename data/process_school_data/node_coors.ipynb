```python
import numpy as np
import pandas as pd

zs_filename = '../zs_weekly/nodes_fixed_age.csv'
nodes_zs = pd.read_csv(zs_filename)
nodes_zs
```

```python
x = []
y = []

trida_map = {
    'A': 1.0,
    'B': 2.0,
    'C': 3.0
}

def random_clipped():
    return np.clip(np.random.normal(scale=0.1), -0.3, 0.3)

for _, r in nodes_zs.iterrows():
    if r["id"] >= 1000:
        trida = r["class"]
        next_x = float(trida[0]) + random_clipped()
        next_y = trida_map[trida[1]] + random_clipped()
    else:
        if r["1. stupeň"] and r["2. stupeň"]:
            next_x = 3
        elif r["2. stupeň"]:
            next_x = 2
        elif r["1. stupeň"]:
            next_x = 1
        else:
            print(r)
            next_x = 4

        next_x += random_clipped()
        next_y = 6 + random_clipped()

    x.append(next_x)
    y.append(next_y)
```

```python
nodes_zs["x"] = x
nodes_zs["y"] = y
```
