```python
import os
import numpy as np
import pandas as pd

zs = True
node_file = './skoly_graph/nodes.csv'
out_dir = './skoly_graph/nodes_with_class_groups/'
seed = 42
n_files = 5
```

```python
if not os.path.exists(out_dir):
    os.mkdir(out_dir)

node_df = pd.read_csv(node_file, index_col=0)
node_df
```

```python
students = (node_df["id"] >= 1000) if zs else (node_df["id"] < 1000)
students = node_df[students]
students
```

```python
np.random.seed(seed)

for i in range(n_files):
    new_df = node_df.copy()
    students = (node_df["id"] >= 1000) if zs else (node_df["id"] < 1000)
    students = node_df[students]

    for student_class in sorted(students["class"].unique()):
        print()
        print(student_class)

        class_mask = students["class"] == student_class
        n_students = sum(class_mask)
        print(n_students)

        class_groups = np.where(np.arange(n_students) <= (n_students // 2), 0, 1)
        np.random.shuffle(class_groups)

        students.loc[class_mask, ["class_group"]] = class_groups

    new_df.to_csv(os.path.join(out_dir, f"nodes_student_groups_{i}.csv"))
```
